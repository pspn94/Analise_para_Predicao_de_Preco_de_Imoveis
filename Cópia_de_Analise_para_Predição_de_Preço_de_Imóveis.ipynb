##Valores Nulos

Começaremos a verificação dos Valores Nulos em nossa tabela de forma a entender como estão se comportando, se é necessário tratamento e quão relevantes são os dados neles obtidos para nossa análise

```
#Criando função para criação do DataFrame de Valores Nulos
def get_valores_nulos(df):

  # Utilizando a função isna para mostrar valores nulos e depois soma-los
  # Gravando a série na variável valores_nulos_serie
  valores_nulos_serie = df.isna().sum()

  # Gravando tamanho total do DataFrame na variável quantidade_total
  quantidade_total = len(df)
  
  # Calculando porcentagem dos valores nulos em relação a quantidade total
  porcentagem_valores_serie = (valores_nulos_serie/quantidade_total)*100
  
  # Criar dicionário
  valores_nulos_dict = {'Quantidade_de_valores_nulos':valores_nulos_serie,
                        'Porcentagem_dos_valores_nulos':porcentagem_valores_serie}

  #Criando DataFrame
  df_valores_nulos = pd.DataFrame(data=valores_nulos_dict) 

  #Ordenar nossos dados do maior para o menor valor
  df_valores_nulos.sort_values(by=['Quantidade_de_valores_nulos'], ascending=False, inplace=True)

  #Eliminando os valores menores ou iguais a 0 da nossa tabela
  df_valores_nulos = df_valores_nulos [df_valores_nulos > 0]
  df_valores_nulos.dropna(axis=0, inplace=True)

  return df_valores_nulos.copy()
```

```
#Mostrando nosso novo DataFrame Valores Nulos em cada coluna e suas porcentagens

df_valores_nulos = get_valores_nulos(df=df.copy())
df_valores_nulos
```

```
#Criando função para plogatem de gráfico
def get_gbarras (df):

  #Criando um gráfico de Barras da quantidade de valores nulos por colunas
  fig = px.bar(data_frame=df_valores_nulos,x=df_valores_nulos.index, y='Quantidade_de_valores_nulos')

  #Trocando o Título do gráfico
  fig.update_layout(title= "Quantidade de Valores Nulos por Coluna")
  #Trocando Título do Eixo X
  fig.update_xaxes(title="Coluna")
  #Trocando o título do Eixo Y
  fig.update_yaxes(title="Quantidade de Valores Nulos")

  

  return fig
```

A ideia principal é **mostrar um gráfico com as Colunas que existem Valores nulos com suas respectivas Quantidade de Valores Nulos destacadas**. Desta forma entendemos de **forma visual** quais são as colunas que mais expressam Valores Nulos e já podemos **presumir** como será o **tratamento** para cada uma delas.

```
fig = get_gbarras(df=df_valores_nulos.copy())
fig.show()
```

###Tratamento de valores nulos

O principal objetivo é **dissecar todas as colunas** para análise e **indicar** e **aplicar** o **melhor tratamento** para elas dentro do DataFrame

```
#Criando função para plotagem de gráfico
def get_gbarras_coluna (df, nome_coluna, eixox, eixoy):

  counts_serie = df[nome_coluna].value_counts()

  fig = px.bar(data_frame=counts_serie,
              x=counts_serie.index,
              y=counts_serie.values)

  fig.update_layout(title=nome_coluna)
  fig.update_xaxes(title=eixox)
  fig.update_yaxes(title=eixoy)


  return fig
```

```
def df_pool_qc (df,nome_coluna):

  # Mostrando se verdadeiro ou falso que os valores da Coluna sao nulos
  # (Verdadeiro quando for nulo / Falso existir valor)
  # Retirando os valores nao nulos
  # Usando o filtro para coluna ['SalePrice'] e gravando a média em media_valor_nulo 
  media_valor_nulo=df[df[nome_coluna].isna()]['SalePrice'].mean()

  # Retirando os valores nulos da Coluna,
  # Calculando a média dos preços da coluna ['SalesPrice'] desse novo database
  # Gravando na variável media_valor_nao_nulo
  media_valor_nao_nulo=df.dropna(axis=0,subset=[nome_coluna],how='any')['SalePrice'].mean()

  # Criando as séries desejadas para o gráfico e atribuindo os valores a elaso 
  media_valor_nulo_e_nao_nulo_serie = [media_valor_nulo, media_valor_nao_nulo]
  string_nulo_e_nao_nulo = ['Valores Nulos', 'Valores Não Nulos']
  
  # Criando novo dicionário 
  medias_nulos_dict = {'Médias_de_Preço':media_valor_nulo_e_nao_nulo_serie,
                        'Valores_Nulos_e_Valores_não_Nulos':string_nulo_e_nao_nulo}

  #Criando DataFrame
  df_medias_nulos = pd.DataFrame(data=medias_nulos_dict) 

  #### Plotando gráfico do Valores Nulos e Não Nulos e seus valores
  fig2 = px.bar(data_frame=df_medias_nulos.copy(),x='Valores_Nulos_e_Valores_não_Nulos', y='Médias_de_Preço')
  fig2.update_layout(title='Comparação de média de preços para Valores Nulos e não Nulos')


  # Retirando os valores nulos do nosso database principal pela Coluna definida pela função
  df.dropna(axis=0,subset=[nome_coluna],how='any',inplace=True)
  # Ordenando pelo novo database pelo ID
  df.sort_values(by='Id',ascending=True, inplace=True, ignore_index=True)

  # Classificando Id como string para melhor apresentação do gráfico
  df['Id']=df['Id'].astype('string')
   
  # Ordenando do maior preço para o menor
  df.sort_values(by='SalePrice',ascending=False,inplace=True)
  
  # Plotando gráfico do Id pelo preço
  fig = px.bar(data_frame=df.copy(),x='Id', y='SalePrice')
  fig.update_layout(title='Análise dos preços para variáveis não nulas')


  fig.show()
  
  fig2.show()

  
```

```
#Criando função para criação do novo DataFrame
def get_valores_zero (df):

  #Criando condição para criação de nova série que mostra os valores iguais a zero na tabela
  df_valores_zero = df==0
  
  #Gravando nova série em valores_zer_soma_serie e somando esta série
  valores_zero_soma_serie=df_valores_zero.sum()
  #Ordenando a série do maior valor para o menor
  valores_zero_soma_serie.sort_values(ascending=False,inplace=True)
  
  #Obtendo valor de quantidade total de linhas
  quantidade_total=len(df)

  #Eliminando as colunas que não tem valores iguais a 0
  valores_zero_soma_serie=valores_zero_soma_serie[valores_zero_soma_serie>0]
  
  #Calculando porcentagem de Valores Iguais a 0 na série
  porcentagem_valores_zero=(valores_zero_soma_serie/quantidade_total)*100
  
  #Criando Dicionário com Valores Iguais a 0 e respectiva Porcentagem deles
  valores_zero_dict = {'Valores_Iguais_a_0':valores_zero_soma_serie,
                      'Porcentagem_dos_valores_iguais_a_zero':porcentagem_valores_zero}
  #Criando um DataFrame
  df_valores_zero = pd.DataFrame(data=valores_zero_dict)
  

  return df_valores_zero.copy()
```

```
#Mostrando Novo DataFrame Valores iguais a 0 e suas porcentagens
df_valores_zero = get_valores_zero(df=df.copy())
df_valores_zero
```

```
#Criando função para plotagem de gráfico
def get_gbarras_valores_zero (df):

  #Criando um gráfico dos Valores Iguais a Zero e suas Porcentagens
  fig = px.bar(data_frame=df_valores_zero, x=df_valores_zero.index,
              y='Porcentagem_dos_valores_iguais_a_zero')
  
  #Trocando o Título do gráfico
  fig.update_layout(title='Quantidade de Valores iguais a zero por Coluna')
  #Trocando Título do Eixo X
  fig.update_xaxes(title='Coluna')
  #Trocando o título do Eixo Y
  fig.update_yaxes(title='Porcentagem de Valores iguais a zero')


  return fig
```

```
#Mostrando o gráfico de Coluna por Porcentagem de Valores Iguais a zero
get_gbarras_valores_zero(df=df.copy())
```

```
def get_valores_zero_por_preco (df,nome_coluna):

  # Criando df com valor bool de igual a 0
  # Aplicando Filtro da coluna 'PoolArea'
  # Aplicando Filtro do Preço
  # Calculando a média
  df_valores_zero = df==0
  media_igual_zero=df[df_valores_zero['PoolArea']]['SalePrice'].mean()

  # Criando df com valor bool diferente de 0
  # Aplicando Filtro da coluna 'PoolArea'
  # Aplicando Filtro do Preço
  # Calculando a média
  df_valores_zero = df!=0
  media_diferente_zero=df[df_valores_zero['PoolArea']]['SalePrice'].mean()

  # Criando as séries desejadas para o gráfico e atribuindo os valores a elas
  
  media_igual_e_diferente_zero= [media_igual_zero,media_diferente_zero]
  media_igual_e_diferente_zero
  string_igual_diferente =['Valor Igual a 0','Valor Diferente de 0']

  # Criando novo dicionário

  media_igual_diferente_dict = {'Médias_de_Preço':media_igual_e_diferente_zero,
                                'Valores_Iguais_e_Diferentes_de_0':string_igual_diferente} 
  
  # Criando DataFrame

  df_medias_zero = pd.DataFrame(data=media_igual_diferente_dict)

  # Plotando gráfico do Valores iguais e diferentes de 0 e seus valores

  fig = px.bar(data_frame=df_medias_zero,x='Valores_Iguais_e_Diferentes_de_0',y='Médias_de_Preço')
  fig.update_layout(title='Comparação de média de preços para Valores iguais e diferentes de 0')

  df_valores_zero = df!=0
  df_filtrado=df[df_valores_zero[nome_coluna]].sort_values(by='Id',
                                                     ascending=True, 
                                                     ignore_index=True)

  # Classificando Id como string para melhor apresentação do gráfico
  df_filtrado['Id']=df_filtrado['Id'].astype('string')
  

  # Ordenando do maior preço para o menor
  df_filtrado.sort_values(by='SalePrice',ascending=False,inplace=True)

  # Plotando gráfico do Id pelo preço
  fig2 = px.bar(data_frame=df_filtrado.copy(),x='Id', y='SalePrice')
  fig2.update_layout(title='Valores diferentes de zero com seu respectivo Id por Preço de Venda')
  
  fig.show()
  fig2.show()

  
```
